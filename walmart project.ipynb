```python
def impute_unitprice(cols):
    unit = cols[0]
    cat = cols[1]
    
    if pd.isnull(unit):

        if cat == 'Home and lifestyle':
            return 50.48

        elif cat == 'Fashion accessories' :
            return 50.10

    else:
        return unit
```

```python
def impute_quantity(cols):
    qua = cols[0]
    cat = cols[1]
    
    if pd.isnull(qua):

        if cat == 'Home and lifestyle':
            return 2

        elif cat == 'Fashion accessories':
            return 2

    else:
        return qua
```

```python
df['unit_price']=df[['unit_price','category']].apply(impute_unitprice,axis=1)
```

```python
df['quantity']=df[['quantity','category']].apply(impute_quantity,axis=1)
```
